```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

dataset = sns.load_dataset('titanic')

dataset.head()
```

```python
print("Shape of dataset:", dataset.shape)
print("\nColumn Names:\n", dataset.columns)
print("\nDataset Info:")
dataset.info()

print("\nStatistical Summary:")
dataset.describe()
```

```python
sns.countplot(x='sex', data=dataset)
plt.title("Count of Male vs Female Passengers")
plt.show()


sns.countplot(x='survived', data=dataset)
plt.title("Survival Count")
plt.show()


sns.barplot(x='sex', y='age', data=dataset)
plt.title("Average Age by Gender")
plt.show()


sns.boxplot(x='sex', y='age', data=dataset)
plt.title("Age Distribution by Gender")
plt.show()


sns.violinplot(x='sex', y='age', hue='survived', data=dataset)
plt.title("Age Distribution with Survival")
plt.show()
```

```python
sns.histplot(dataset['age'], bins=10, kde=True)
plt.title("Age Distribution")
plt.show()


sns.jointplot(x='age', y='fare', data=dataset, kind='scatter')
plt.show()


sns.rugplot(dataset['fare'])
plt.title("Rug Plot for Fare")
plt.show()
```

```python
corr = dataset.corr(numeric_only=True)

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.histplot(dataset['fare'], bins=10, kde=False)

plt.title("Distribution of Ticket Fare")
plt.xlabel("Fare")
plt.ylabel("Frequency")

plt.show()
```
