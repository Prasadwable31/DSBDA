```python
import nltk

nltk_data_path = "C:/Users/harsh/nltk_data"
nltk.data.path.append(nltk_data_path)

# Download required datasets
nltk.download('punkt', download_dir=nltk_data_path)
nltk.download('punkt_tab', download_dir=nltk_data_path)
nltk.download('stopwords', download_dir=nltk_data_path)
nltk.download('wordnet', download_dir=nltk_data_path)
nltk.download('averaged_perceptron_tagger', download_dir=nltk_data_path)
nltk.download('averaged_perceptron_tagger_eng', download_dir=nltk_data_path)
```

```python
text = """Text analytics is very important in today's world. 
Companies like Amazon and YouTube analyze user reviews to understand feedback. 
Machine learning and NLP help in processing textual data efficiently."""
```

```python
from nltk.tokenize import wordpunct_tokenize

word_tokens = wordpunct_tokenize(text)
print("Word Tokens:\n", word_tokens)
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

filtered_words = [w for w in word_tokens if w.lower() not in stop_words]

print("Filtered Words:\n", filtered_words)
```

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

stemmed_words = [ps.stem(w) for w in filtered_words]

print("Stemmed Words:\n", stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(w) for w in filtered_words]

print("Lemmatized Words:\n", lemmatized_words)
```

```python
pos_tags = nltk.pos_tag(word_tokens)
print("POS Tags:\n", pos_tags)
```

```python
import pandas as pd

data = {
    'text': [
        "I love machine learning",
        "Machine learning is amazing",
        "I love coding in Python",
        "Python is great for data science"
    ],
    'label': [1, 1, 0, 0]
}

df = pd.DataFrame(data)
print(df)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

X = tfidf.fit_transform(df['text'])
y = df['label']

print("TF-IDF Matrix:\n", X.toarray())
print("Features:\n", tfidf.get_feature_names_out())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics

model = MultinomialNB()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Predictions:", y_pred)
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```
